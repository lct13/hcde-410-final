## Part 1: Cleaning the license transaction data

```python
#First I'm importing the csv module, a little code toolkit for working with spreadsheet-style data files
import csv
```

### 1.1 Load the data into the notebook

According to the documetation, the data for professional licenses in WA from July 2020 until now is contained in ``Professional_License_Transactions_by_Department_of_Licensing.tsv`` ([documentation](https://data.wa.gov/Employment/Professional-License-Transactions-by-Department-of/ixni-jq78)).

The function below will load in a tab-separated (.tsv) file and convert it into a lists-of-dictionaries. 

```python
def prepare_datasets(file_path):
    """ 
    Accepts: path to a tab-separated plaintext file
    Returns: a list containing a dictionary for every row in the file, 
        with the file column headers as keys
    """
    
    with open(file_path) as infile:
        reader = csv.DictReader(infile, delimiter='\t')
        list_of_dicts = [dict(r) for r in reader]
        
    return list_of_dicts
```

### 1.2 Understanding the fields within the dataset
Let's load the three .tsv files into Python and store them as a variable that can be referenced to later on.

Once we've created this list-of-dicts, we will do two things to check our work so far: 
- we will print the first transaction data (list index ``[0]``) so that we know what the format looks like
- we will print the length of the list, divided by 34 months, to see how many transactions we have per month on average (it should be 34 months from July 2020 - April 2023)

```python
#load the data from the flat files into three lists-of-dictionaries
license_transactions = prepare_datasets("Professional_License_Transactions_by_Department_of_Licensing.tsv") #Wikipedia_Talk_Labels_Toxicity_4563973/
print(license_transactions[0])
print("total: " + str(len(license_transactions)))
#let's see how many transactions we have per month on average (it should be 34 months from July 2020 - April 2023)
print("average per month: " + str(len(license_transactions)/34))
```

### 1.3 Checking for duplicate rows within the dataset
This function comepares each row to the rest in the dataset, and prints a row if it has an exact match.
Note: This will take a while to run since it's a nested loop, so I terminated it after finding 5 duplicates (about 1 minute). This is still good information to look at.

```python
def has_dupes(transaction_list):
    """
    Accepts: a list of dictionaries 
    
    If a duplicate row is found, 
        print the complete dictionary of transaction data for that row
    
    """
    # switch out `1000` with `len(transaction_list)` to run through the whole dataset
    for i in range(0, 1000): #len(transaction_list)):
        for j in range(i+1, 1000): #len(transaction_list)):
            if transaction_list[i] == transaction_list[j]:
                print(transaction_list[i])
    print("done!")
            
has_dupes(license_transactions);
```

## Part 2: Analyzing licenses distribution in graphs

Now that we have gained an understanding of the dataset, we can start using it to ask and answer research questions! 
First, let's look at overall distribution of all vs. tattoo vs. nail licenses, within WA state.

Since I want to analyze the data based on different categories like county within WA, state, transaction month, I'll write a general function for it. This function takes in the dataset and the category to count the data by, and return a dictionary of counts of licenses based on the specified category. It also keeps track of the total counts.
Let's print out the totals to make sure everything looks right.

```python
def distribution_by(data, category): #eg. data=all_annotators, category='County', licenses=['Tattoo Artist', Manicurist']
    dist = {}
    totals = {'all':0, 'tattoo':0, 'nail':0}
    for row in data:
        if (not row[category] in dist.keys()):
            dist[row[category]] = {'all':0, 'tattoo':0, 'nail':0}
        dist[row[category]]['all'] += 1
        totals['all'] += 1
        if row['License Type'] == 'Tattoo Artist':
            dist[row[category]]['tattoo'] += 1
            totals['tattoo'] += 1
        if row['License Type'] == 'Manicurist':
            dist[row[category]]['nail'] += 1
            totals['nail'] += 1
    dist['totals'] = totals
    return dist

dist_by_county = distribution_by(license_transactions, 'County')
print(dist_by_county['totals'])
```

That looks reasonable, I'd expect the total counts for tattoo and nail to be close. The total for all licenses was also 50216 from the first step, so we're all set to start making tables and graphs!

### 2.1 Distribution of licenses across WA state
I want to answer the following questions
- **Q1.** What is the distribution of these licenses by county within the state of Washington?
- **Q2.** What is the distribution of the licenses by state, if out of state?

I'm using the [Pandas](https://pandas.pydata.org/pandas-docs/version/0.15/10min.html) library to convert data into tidy tabular representations. I'll also use [Plotly](https://plotly.com/python/v3/statistics-charts/#new-to-plotly) to create charts.

```python
import pandas as pd
def display_dist(name, dist_dict):
    with open(name + '.csv', 'w', encoding='utf-8') as f:
        writer = csv.writer(f)
        #write a header row
        writer.writerow(('county_from',
                         'tattoo_artists', 
                         'manicurists',
                         'all licenses',))
        #loop through our dataset and write it to the file, row by row
        for k in dist_dict.keys():
            if k != 'totals':
                writer.writerow((k, 
                                 100*dist_dict[k]['tattoo']/dist_dict['totals']['tattoo'],
                                 100*dist_dict[k]['nail']/ dist_dict['totals']['nail'],
                                 100*dist_dict[k]['all']/ dist_dict['totals']['all']))        
    
display_dist('Distribution_TattooVsNail_byCounty', dist_by_county)
df_plot = pd.read_csv('Distribution_TattooVsNail_byCounty.csv')
df_plot
```

```python
import matplotlib.pyplot as plt
df_plot = pd.read_csv('Distribution_TattooVsNail_byCounty.csv')

# Create plot
plt.subplots(figsize=(12, 7))
plt.plot(df_plot['county_from'], df_plot.iloc[:, 1:], label = df_plot.columns[1:])

# Customize plot
plt.title("Professional license transactions in WA from July 2020 - May 2023", size = 15)
plt.xlabel("County", size = 12)
plt.ylabel("Percentage %", size = 12)
plt.xticks(rotation=45)
plt.legend()

# Display plot
plt.show()
```

**Q1.** What is the distribution of these licenses by county within the state of Washington?

This is interesting: 
* All 3 graphs follow a similar patterns in terms of where they peak, even when it's at different rate/slope. Seems like populated areas, with more opportunities for business would have higher percentages of licenses with not much anomality. The percantages of manicurist licenses do match up with all licenses more so than tattooist ones. Tattoos might just be even less popular in areas with less businesses.
* The percentage of manicurists from out of state is far higher than that of tattoo artists (around 2.6 times higher), while the percentage for all licenses fall in the middle. This does make sense, not a lot of people would move out of state to be a tattooist. In fact, of all the professional tattooists I've met in Seattle (as a learning artist), they are all from Washington. Personally I wouldn't stay in Seattle to do tattoos either, I'm just here for school. To test this assumption about manicurist licenses, let's take a look at "State" data.

Moving on to the next question: 

**Q2.** What is the distribution of the licenses by state, if out of state?

First I'll make a dictionary of license counts, using the previous function I wrote. Then I'll put this in a csv, but I only want to look at popular states. So let's try cut off at 0.5%?
I'll use permille instead of percentage, so the numbers look prettier for the graphs. I'm also leaving out WA, since I'm only looking at out of state.

```python
dist_by_state = distribution_by(license_transactions, 'State')
cutoff = 5;

with open('Distribution_TattooVsNail_byState.csv', 'w', encoding='utf-8') as f:
    writer = csv.writer(f)
    #write a header row
    writer.writerow(('state_from',
                     'tattoo_artists', 
                     'manicurists',
                     'all licenses',))
    #loop through our dataset and write it to the file, row by row
    for k in dist_by_state.keys():
        # skip the row for WA, and only select states with > 5 permilles of nail licenses
        if k != 'totals' and k!= 'WA' and 1000*dist_by_state[k]['nail']/ dist_by_state['totals']['nail'] > cutoff:
            writer.writerow((k, 
                             1000*dist_by_state[k]['tattoo']/dist_by_state['totals']['tattoo'],
                             1000*dist_by_state[k]['nail']/ dist_by_state['totals']['nail'],
                             1000*dist_by_state[k]['all']/ dist_by_state['totals']['all']))  
df_plot = pd.read_csv('Distribution_TattooVsNail_byState.csv')
df_plot
```

```python
# Create plot
plt.subplots(figsize=(12, 7))
plt.plot(df_plot['state_from'], df_plot.iloc[:, 1:], label = df_plot.columns[1:])

# Customize plot
plt.title("Professional license transactions in WA from July 2020 - May 2023", size = 15)
plt.xlabel("State", size = 12)
plt.ylabel("Permille %o", size = 12)
plt.xticks(rotation=45)
plt.legend()

# Display plot
plt.show()
```

* So, looks like the permilles of manicurist licenses are generally higher than those of all licenses, but we already know that (from the county data: out of state for manicurist is 37.5%, compared to 26.5% for all). States that seem to stand out, with peaks that don't correspond with all licenses, are TX, IN, NE, and MI. Only TX and NE can be considered states with high Vietnamese populations (see secreenshot below from [usa.com](http://www.usa.com/rank/us--vietnamese-population-percentage--state-rank.htm). Thus, hypothesis 1b) *There are more out of state licenses for manicure(nail), and within those there are more from states with high Vietnamese populations* is not strongly supported.

* Interesting peak in tattoo licenses from Oregon (54% compared to 30% for all licenses), but I'd think it makes sense. A lot of people I know would travel to Oregon (from WA) to get tattoos, so there  must be a large population of tattoo artists there. Portland is a quirky city. Maybe a lot of tattooists from OR move to WA for more business. This is an unexpected finding, exciting!



### 2.2 Distribution of license transactions over time
Let's look at the distributions over time now. I'll add up all the transactions for each month in the database, and put that on the graph.

Here I'm looking at actual counts instead of percentages, and I'll divide the counts for *overall licenses* by 10 so they show up nicely on the graph, with trends I can still identify.

```python
dist_by_month = distribution_by(license_transactions, '\ufeffTransaction Month')

with open('Distribution_TattooVsNail_byMonth.csv', 'w', encoding='utf-8') as f:
    writer = csv.writer(f)
    #write a header row
    writer.writerow(('month',
                     'tattoo_artists', 
                     'manicurists',
                     'all licenses/10',))
    #loop through our dataset and write it to the file, row by row
    for k in dist_by_month.keys():
        if k != 'totals':
            writer.writerow((k,
                             dist_by_month[k]['tattoo'], dist_by_month[k]['nail'], dist_by_month[k]['all']/10))
                             # 1000*dist_by_month[k]['tattoo']/dist_by_month['totals']['tattoo'],
                             # 1000*dist_by_month[k]['nail']/ dist_by_month['totals']['nail'],
                             # 1000*dist_by_month[k]['all']/ dist_by_month['totals']['all']))  
df_plot = pd.read_csv('Distribution_TattooVsNail_byMonth.csv')
df_plot
```

```python
# Create plot
plt.subplots(figsize=(12, 7))
plt.plot(df_plot['month'], df_plot.iloc[:, 1:], label = df_plot.columns[1:])

# Customize plot
plt.title("Professional license transactions in WA from July 2020 - May 2023", size = 15)
plt.xlabel("Month", size = 12)
plt.ylabel("Count", size = 12)
plt.xticks(rotation=45)
plt.legend()

# Display plot
plt.show()
```

Great! The dataset is already sorted in order of time, so I don't have to do extra work to get a linear presentation of time.

* Both counts for manicurists and overall licenses start lower and picks up gradually over the last few months of 2020 (July-December), but counts of tattoo artists stayed neutral. This period of time was when Covid regulations started relaxing just a little bit, and I remember my auntie's nail salon was starting to operate closer-to-normal hours and get business again. It's interesting that counts of tattoo license transactions stayed the same, maybe Covid regulations didn't affect business for tattooists as much. I remember most tattoo shops would accept customers through out the pandemic but required masks and extra sanitization. Most of the shops I follow closely still keep their mask policies until today.

* **Most noticeable was the steep drop in 2021 November - 2022 February**. This does correspond with the discovery of the Omicron variant of Covid. But there could have been some other event since I don't remember a lot of places closing up, it was mostly reinforcement of masks and vaccines. Count of people getting licenses seem to pick right back up in March, so it didn't last long.

* License transactions for tattoo and all licenses, but **not** for manicurist, pick up slightly in August 2022, declines slightly for the last couple months and suddenly picks up again in January 2023. I wonder if there's an economic event or if it's just seasonal depression.

Overall, there are trends that can trace back to Covid-related events, but no way to be 100% sure.
